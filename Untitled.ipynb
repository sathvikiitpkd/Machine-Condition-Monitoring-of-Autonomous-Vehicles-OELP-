```python
import librosa.display, os
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.utils import load_img, img_to_array
def load_images_from_path(path, label):
    images = []
    labels = []

    for file in os.listdir(path):
        images.append(img_to_array(load_img(os.path.join(path, file), target_size=(224, 224, 3))))
        labels.append((label))
        
    return images, labels

def show_images(images):
    fig, axes = plt.subplots(1, 12, figsize=(20, 20), subplot_kw={'xticks': [], 'yticks': []})

    for i, ax in enumerate(axes.flat):
        ax.imshow(images[i] / 255)
        
x = []
y = []
```

```python
images, labels = load_images_from_path('./Spectrograms/Healthy', 0)
show_images(images)
    
x += images
y += labels
```

```python
images, labels = load_images_from_path('./Spectrograms/Ball_defect', 1)
show_images(images)
    
x += images
y += labels
```

```python
images, labels = load_images_from_path('./Spectrograms/Inner_race_defect', 2)
show_images(images)
    
x += images
y += labels
```

```python
images, labels = load_images_from_path('./Spectrograms/Outer_race_defect', 3)
show_images(images)
    
x += images
y += labels
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.3, random_state=5)

x_train_norm = np.array(x_train) / 255
x_test_norm = np.array(x_test) / 255
print(x_train_norm.shape)
y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout
from keras.layers import Flatten, Dense
from tensorflow.keras import layers

# model
model = Sequential([
    layers.Conv2D(filters = 64, kernel_size = 5,activation= 'LeakyReLU', input_shape = (224,224,3)),
    layers.MaxPool2D(2),
    layers.Flatten(),
    layers.Dense(120,activation = 'LeakyReLU'),
    layers.Dense(84, activation = 'LeakyReLU'),
    layers.Dense(4, activation = 'softmax')
])
model.summary()
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics= ['accuracy'])
```

```python
hist = model.fit(x_train_norm, y_train_encoded, validation_data=(x_test_norm, y_test_encoded), batch_size=10, epochs=25)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
sns.set()

y_predicted = model.predict(x_test_norm)
mat = confusion_matrix(y_test_encoded.argmax(axis=1), y_predicted.argmax(axis=1))
class_labels = ['Healthy', 'Ball_defect', 'Inner_race_defect', 'Outer_race_defect']

sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False, cmap='Blues',
            xticklabels=class_labels,
            yticklabels=class_labels)

plt.xlabel('Predicted label')
plt.ylabel('Actual label')
```
